```python
import numpy as np
from numpy import genfromtxt
import librosa
import pandas as pd
```

```python
df_1_arr = genfromtxt(r'C:\Users\Admin\Desktop\DS203\E7\MFCC-files-v2\01-MFCC.csv', delimiter=',')
df_1_arr.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample array for visualization
# mfcc_array = np.random.rand(20, 9769)  # Uncomment if you want to test with random data

# Plotting the MFCCs as a heatmap
plt.figure(figsize=(15, 6))
plt.imshow(df_1_arr, aspect='auto', cmap='viridis')
plt.colorbar(label='MFCC Coefficient Value')
plt.xlabel('Time Frame')
plt.ylabel('MFCC Coefficient')
plt.title('MFCC Coefficients Heatmap')
plt.show()
```

```python
# Plotting selected MFCC coefficients over time
plt.figure(figsize=(12, 6))
for i in range(20):  # Plot the first 5 coefficients as an example
    plt.plot(df_1_arr[i, :], label=f'MFCC Coefficient {i+1}')
plt.xlabel('Time Frame')
plt.ylabel('Coefficient Value')
plt.title('Selected MFCC Coefficients over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (14, 4))
librosa.display.specshow(df_1_arr, x_axis = 'time', sr=44100)
plt.colorbar(format='%+2.f')
plt.title('MFCCs')
```

```python
delta_mfccs = librosa.feature.delta(df_1_arr)
delta2_mfccs = librosa.feature.delta(df_1_arr, order=2)
```

```python
plt.figure(figsize = (14, 4))
librosa.display.specshow(delta_mfccs, x_axis = 'time', sr=44100)
plt.colorbar(format='%+2.f')
plt.title('MFCCs')
```

```python
plt.figure(figsize = (14, 4))
librosa.display.specshow(delta2_mfccs, x_axis = 'time', sr=44100)
plt.colorbar(format='%+2.f')
plt.title('MFCCs')
```
