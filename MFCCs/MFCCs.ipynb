```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd
```

```python
violin_c, sr = librosa.load('./violin_c.wav')

plt.figure(figsize = (14, 4))
librosa.display.waveshow(violin_c, alpha = 0.5)
plt.title('Violin_C')
```

```python
mfcc = librosa.feature.mfcc(y=violin_c, sr=sr, n_mfcc=13)
mfcc.shape
```

```python
plt.figure(figsize = (14, 4))
librosa.display.specshow(mfcc, x_axis = 'time', sr=sr)
plt.colorbar(format='%+2.f')
plt.title('MFCCs')
```

```python
delta_mfccs = librosa.feature.delta(mfcc)
delta2_mfccs = librosa.feature.delta(mfcc, order=2)

print(f"Shape of Delta MFCCs: {delta_mfccs.shape}")
print(f"Shape of Delta_2 MFCCs: {delta2_mfccs.shape}")
```

```python
plt.figure(figsize = (14, 4))
librosa.display.specshow(delta_mfccs, sr=sr, x_axis='time')
plt.colorbar(format='%+2.f')
plt.title('Delta MFFCs')
```

```python
plt.figure(figsize=(14, 4))
librosa.display.specshow(delta2_mfccs, sr=sr, x_axis='time')
plt.colorbar(format='%+2.f')
plt.title('Delta 2 MFCCs')
```
