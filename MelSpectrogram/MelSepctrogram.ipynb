```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd
```

```python
violin_c, sr = librosa.load('./violin_c.wav')

plt.figure(figsize=(10, 5))
librosa.display.waveshow(violin_c, alpha=0.5)
plt.title('Waveform of Violin_C')
```

```python
filter_banks = librosa.filters.mel(sr=sr, n_fft = 2048, n_mels = 10)
filter_banks.shape
```

```python
plt.figure(figsize=(10, 6))
librosa.display.specshow(filter_banks, sr=sr, x_axis = 'linear')
plt.colorbar(format='%+2.f')
```

```python
mel_spectrogram = librosa.feature.melspectrogram(y=violin_c, sr=sr, n_fft=2048, hop_length=512, n_mels=10)
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
log_mel_spectrogram.shape
```

```python
plt.figure(figsize = (10, 6))
librosa.display.specshow(log_mel_spectrogram, x_axis='time', y_axis='mel', sr=sr)
plt.colorbar(format='%+2.f')
plt.title('Mel-Spectrogram')
```
